```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv('fruit_dataset.csv')

df.info()
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X=df[['mass','width','height']]
Y=df['fruit_label']
X
```

```python
Y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size = .20,random_state=0)
X_train.shape  #(47,3)
X_test.shape   #(12,3)
y_train.shape  #(47,)
y_test.shape   #(12,)
```

```python
knn=KNeighborsClassifier(n_neighbors=5, metric='minkowski')

knn.fit(X_train,y_train)
```

```python
import numpy as np

y_pred = knn.predict(X_test)
y_pred
```

**#Lets make predictions with new data:**

```python
prediction1=knn.predict([['100','6.3','8']])  # new mass, width,height
prediction1    #output is lemon (4)
```

```python
prediction2=knn.predict([['300','7','10']])
prediction2   #output is orange (3)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy is",accuracy_score(y_test, y_pred))
```
